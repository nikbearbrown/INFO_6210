```python
import os
import pandas as pd
from glob import glob

combined_csv = pd.concat( [ pd.read_csv(f) for f in glob('Glassdoor_Ratings\\*.csv') ])
                                                
```

```python
#Splitting the file into two tables GLASSDOOR_RATINGS & GLASSDOOR_REVIEWS

GlassdoorFile = pd.read_csv("combined_csv_glassdoor.csv")

glassdoor_Reviews = GlassdoorFile[['Company_ID', 'Review_ID', 'date', 'employee_title', 'location', 'employee_status', 'review_title', 'years_at_company', 'helpful', 'pros', 'cons', 'advice_to_mgmt']]
glassdoor_Reviews.to_csv('GLASSDOOR_REVIEWS.csv')
```

```python
#GLASSDOOR_RATINGS

glassdoor_Ratings = GlassdoorFile[['Company_ID', 'Review_ID', 'rating_overall', 'rating_balance', 'rating_culture', 'rating_career', 'rating_comp', 'rating_mgmt']]
glassdoor_Ratings.to_csv('GLASSDOOR_RATINGS.csv')
```

```python
#splitting youtube data into YOUTUBE_DATA and YOUTUBE_VIDEO_DATA

youtube_data = pd.read_csv('youtube_complete_data.csv')

youtube_Data = youtube_data[['channelId', 'channelTitle', 'categoryId']]
youtube_Data.to_csv('YOUTUBE_DATA.csv', index = None)

youtube_Video_Data = youtube_data[['Company_Id', 'videoId', 'viewCount', 'commentCount', 'likeCount', 'dislikeCount', 'channelId', 'tags', 'publishedAt']]
youtube_Video_Data.to_csv('YOUTUBE_VIDEO_DATA.csv', index = None)
```

```python
#splitting tweets data into TWEETS_TBL nad TWEETS_USER_DATA_TBL

tweets_data = pd.read_csv('tweet_merge_Data.csv')
tweets_data.head()

tweets_Data = tweets_data[['Company_Id','user_id','tweet_id','text', 'retweet_count', 'favorite_count', 'hashtags']]
tweets_Data.to_csv('TWEETS_TBL.csv', index = None)


tweets_UserData = tweets_data[['user_id','user_name', 'user_screen_name', 'user_followers_count', 'user_location']]
tweets_UserData.to_csv('TWEETS_USER_DATA_TBL.csv', index = None)
```
