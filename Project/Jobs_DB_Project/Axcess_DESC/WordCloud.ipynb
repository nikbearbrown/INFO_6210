```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
import collections
import re
from os import path
from PIL import Image
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import math

from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import nltk
nltk.download('stopwords')
```

```python
with open("url_complete_data_Axcess.txt") as file_content:
     word_list  = [word.lower() for line in file_content for word in re.findall(r'\w+', line)]
        
filtered_words = list(filter(lambda word: word not in stopwords.words('english'), word_list))
filtered_words2 = list(filter(lambda word: word not in ['axcess','financial','want','need','way', 'rich', 'within', 'sure', 'required', 'little', 'needs', 'life'], filtered_words))
```

```python
wordcount = collections.defaultdict(int)
for word in filtered_words2:
        wordcount[word] += 1
```

```python
mc = sorted(wordcount.items(), key=lambda k_v: k_v[1], reverse=True)

mc = dict(mc)
words_c = list(mc.keys())
values = list(mc.values())

wordcloud = WordCloud(stopwords=STOPWORDS,
                            background_color = 'white',
                            width=1200,height=1000,
                            max_words=700,
                            random_state=42,
                            repeat = False,
                            relative_scaling='auto'
                            ).generate(str(words_c))

plt.figure(figsize = (11, 11), facecolor = None) 
plt.imshow(wordcloud) 
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
