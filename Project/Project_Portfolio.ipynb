You have a choice to contribute to the jobs database or to contribute to a "Bantaba" project related to databases. 


### Jobs Database 

In this project students will contribute to building a jobs database using mysql and python with proper documentation. Students will each focus on job domains like computational biology jobs, finance jobs, game analytics, game AI, data visualization jobs, Tableau jobs, Northeastern jobs, small lab jobs, etc..  
The database is focused around an employer like a company and research lab. Each student should collect information on at least 300 employers.    

_The Project_

You have been given the beginnings of a jobs database but the previous developer was let go because the organization of the previous project is a mess. The developers all worked independently and the tables and schemas didn’t match. There are some working scripts but they sometimes replicate and some are missing. Some scripts have bugs and may be in python 2.7.  

Data is in .csv, json, sqlite and mysql.

Your job is to salvage what data, schemas and scripts that you can and add what is missing.    

Everything must be done in mysql using the InnoDB engine. 

All python scripst must be in python 3 or above and using Google Python Style Guide http://google.github.io/styleguide/pyguide.html

All scripts and SQL files must have meaningful names.

_Tasks_  

One domain can be work on by groups of up to 3 people.  E-mail the TAs to make sure your domain isn’t taken.  Jobs is too broad for a domain, domains are computational biology jobs, finance jobs, game analytics, game AI, data visualization jobs, Tableau jobs, Northeastern jobs, small lab jobs, etc. It is OK to have some overlap between domains.

One you have a domain do the following tasks:  
1. Build the list of companies in some domain.  
2. Web scrape the data for each of the companies for job details.   
3. Automate the scraping.   
4. Get additional relevant data from sites like Glassdoor, LinkedIn (one per person)   
5. Get data from social media sites – Twitter, YouTube, Instagram, Steam, Twitter, etc.. (One per person)  
6. Tag the social media posts, including synonyms for the tags   
7. Clean and integrate data.   
8. Build an ER diagram and model the db.   
9. Build the dB schema and insert the data   
10. Generate use cases.   
11. Optimize the database.    
12. Properly document that database
13. Professionalism (Licensing, code style, file naming, README. Etc.)
